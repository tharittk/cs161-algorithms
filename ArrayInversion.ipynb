```python
arr = '132546' # ans = 2

def array_inversion_mergeSort(arr):
    n = len(arr)

    # odd case
    if n == 1:
        return arr, 0
    else:
        lhs = arr[ : n//2 ]
        rhs = arr[ n//2 : ]

    # base-case
    if len(rhs) == 1 and len(lhs) == 1:
        # merge
        if int(lhs) <= int(rhs):
            return lhs + rhs, 0
        else:
            return rhs + lhs, 1
    else:
        lhs_out, countLhs = array_inversion_mergeSort(lhs)
        rhs_out, countRhs = array_inversion_mergeSort(rhs)

        totalCount = countLhs + countRhs
        n_out = len(lhs_out) + len(rhs_out)
        sortedArray = ''
        # merge sorted array & counting count = countRhs
        i = 0
        j = 0
        for k in range(n_out):

            if int(lhs_out[i]) <= int(rhs_out[j]):
                sortedArray += lhs_out[i]
                i += 1
            else: # right is smaller
                sortedArray += rhs_out[j]
                j += 1
                totalCount += (len(lhs_out) - i)

            if i == len(lhs_out):
                sortedArray += rhs_out[j:]
                break

            if j == len(rhs_out):
                sortedArray += lhs_out[i:]
                break
        return sortedArray, totalCount

#array_inversion_mergeSort(arr)

```

```python
with open("IntegerArray.txt") as f:
    lines = [line.rstrip('\n') for line in f]

total = 0
for line in lines:
    _arr, count = array_inversion_mergeSort(line)
    print("BF ",line, "AF",_arr , "count: ", count)
    total += count
print(total)
```
